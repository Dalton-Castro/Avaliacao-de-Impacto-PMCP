# <font color='darkblue'>PROGRAMA MÃE CORUJA PERNAMBUCANA (PMCP) </font>
<hr style='border: 2px solid darkblue;'>


### Implementado em 2008, no estado de Pernambuco, o Programa Mãe Coruja Pernambucana tinha como objetivo reduzir os altos índices de mortalidade infantil e materna no estado de Pernambuco.
    
### O PMCP atua por meio de uma extensa rede de ações calcadas em assistências no pré-natal, acompanhamento das crianças e auxílio durante a primeira infância, das quais o amparo ao período gestacional da mulher se destaca como o seu maior enfoque.

### Implementação:

***
#### A escolha da região e dos municípios por onde o programa começaria a ser implantado obedeceu a alguns critérios, sendo levada, ainda, em consideração as onze Gerências Regionais de Saúde (Geres) responsáveis pelos 185 municípios do estado. 

#### Os principais critérios utilizados foram a taxa de mortalidade infantil, os Índice de Desenvolvimento Humano (IDH) e Índice de Desenvolvimento Infantil (IDI). Seriam considerados aqueles municípios que possuíssem uma taxa de mortalidade infantil maior ou igual a 25 óbitos a cada mil nascidos vivos e que apresentassem baixos IDH e IDI. Além disso, determinou-se que o programa seria implantado em todos os municípios das Geres para os casos em que ao menos 50\% dos municípios nela situados  detivessem uma taxa de mortalidade infantil superior a 25 óbitos por mil nascidos vivos.


#### Com base nesses parâmetros, a implementação iniciou-se pela microregião de Araripina (Sertão do Araripe), seguida pelo Sertão do Moxotó (VI Geres). Desse modo, 24 municípios foram contemplados pelo PMCP em sua primeira fase de implementação. Ao longo dos anos o programa se ampliou de forma a atuar, atualmente, em 105 dos 185 municípios do estado.

# <font color='red' style='font-size: 30px;'>Substituindo ',' por '.'  e excluindo linhas nulas nas Proficiências Médias</font>
<hr style='border: 2px solid red;'>

### De início, é perceptível que as variáveis quantitativas do dataset têm suas casas decimais separadas por vírgulas ao invés de pontos. 

### É necessário substituir essa vírgulas por pontos para conseguirmos transformar a coluna em numeric para podermos efetuar futuros cálculos. 

### Inicialmente, há interesse apenas na coluna relativa às proficiências médias, portanto, apenas ela passará por esse procedimento.

```python
proficiencia = []

for item in dados['Proficiência Média']:
    proficiencia.append(item.replace(',','.'))

dados['Proficiência Média'] = proficiencia
dados['Proficiência Média'] = pd.to_numeric(dados['Proficiência Média'])
```

### **O erro acima nos reporta que algumas linhas contém o caractere '-' para reportar valores inexistente.**

**O seguinte código substitui o '-' por NA, nos mermitindo facilmente remover observações nulas dos nossos dados.**

```python
proficiencia = []

for item in dados['Proficiência Média']:
    proficiencia.append(item.replace(',','.').replace('-', ''))

dados['Proficiência Média'] = proficiencia
dados['Proficiência Média'] = pd.to_numeric(dados['Proficiência Média'], errors='coerce')
dados['Proficiência Média'] = dados['Proficiência Média'].astype(float)
```

```python
dados.dropna(subset = ['Proficiência Média'], inplace = True)
```

**Também transformamos a coluna para numeric e em seguida para o tipo float. Como estamos apenas interessados na proficiência média, não repetiremos o processo para as demais colunas.**

# <font color='red' style='font-size: 30px;'>Separando os Grupos de Controle e Tratamento</font>

## Obtendo os Códigos dos Grupos de Tratamento e Controle:

<hr style='border: 2px solid black;'>

```python
participacao_no_PMCP = pd.read_excel('Dados/tabela_de_participacao.xls')
```

```python
participacao_no_PMCP
```

**Nosso grupo de tratamento são as escolas dos municípios que entraram no PMCP em 2008**

**Nosso grupo de controle são as escolas dos demais municípios**

**Para separá-las na nossa base de dados, devemos obter os códigos dos municípios entrantes e não entrantes no PMCP em 2008**

```python
codigos_municipios_tratados = participacao_no_PMCP.query("`2008` == 'SIM'")['Códigos dos Municípios'].reset_index(drop=True)
codigos_municipios_tratados
```

```python
codigos_municipios_controle = participacao_no_PMCP.query("`2008` != 'SIM'")['Códigos dos Municípios'].reset_index(drop=True)
codigos_municipios_controle
```

## Grupo de Tratamento:
<hr style='border: 2px solid black;'>

```python
dados_grupo_de_tratamento = dados[dados['Código do Município'].isin(codigos_municipios_tratados)]
dados_grupo_de_tratamento.reset_index(drop=True, inplace=True)
dados_grupo_de_tratamento.head()
```

## Grupo de Controle
<hr style='border: 2px solid black;'>

```python
dados_grupo_de_controle = dados[dados['Código do Município'].isin(codigos_municipios_controle)]
dados_grupo_de_controle.reset_index(drop=True, inplace=True)
dados_grupo_de_controle.head()
```

# <font color='red' style='font-size: 30px;'>Agregando Por Municípios</font>
<hr style='border: 2px solid red;'>

**Iremos agregar os dados a nível municipal, tirando a média aritmética das proficiências médias das escolas** 

```python
dados_municipios = dados.groupby(['Código do Município', 'Edição'], as_index=False)['Proficiência Média'].mean()
dados_municipios.sort_values(['Código do Município','Edição'], inplace=True)
dados_municipios.rename(columns={'Código do Município': 'codigo_do_municipio', 
                                 'Edição': 'ano', 
                                 'Proficiência Média': 'proficiencia_media_matematica'},inplace=True)
dados_municipios
```

**Checando se todos os municípios possuem dados em todos os anos**

```python
dados_municipios.codigo_do_municipio.value_counts()
```

**O município com o código 2605905 não possui dados da proficiência média para todos os anos**    
**Portanto, será excluído da nossa base de dados**

```python
dados_municipios = dados_municipios.query('codigo_do_municipio != 2605905')
dados_municipios
```

## Grupo de Tratamento:
<hr style='border: 2px solid black;'>

```python
selecao_municipios_tratamento = dados_municipios['codigo_do_municipio'].isin(codigos_municipios_tratados)
selecao_municipios_tratamento
```

```python
dados_municipios_tratamento = dados_municipios[selecao_municipios_tratamento]
dados_municipios_tratamento.reset_index(drop=True, inplace=True)
dados_municipios_tratamento
```

## Grupo de Controle:
<hr style='border: 2px solid black;'>

```python
selecao_municipios_controle = dados_municipios['codigo_do_municipio'].isin(codigos_municipios_controle)
selecao_municipios_controle
```

```python
dados_municipios_controle = dados_municipios[selecao_municipios_controle]
dados_municipios_controle.reset_index(drop=True, inplace=True)
```
